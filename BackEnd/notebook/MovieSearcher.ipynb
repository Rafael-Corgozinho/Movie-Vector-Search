### Importing the Dataset

Here, we use the Kaggle dataset path to download and access the dataset.  
Since the returned path points to the dataset directory rather than the CSV file itself, we need to locate the CSV file inside the directory before loading it with pandas.

```python
import kagglehub
import pandas as pd
import os
import chromadb

# Download dataset
path = kagglehub.dataset_download("shivamb/netflix-shows")

# Ver arquivos
print(os.listdir(path))

# Caminho do CSV
csv_path = os.path.join(path, "netflix_titles.csv")
df = pd.read_csv(csv_path)

df.head()
```

### Pre-Processing

As we need to run the embendding in the string, we transformed every column that we need into a unique collumn called Overview so we can transform it.

```python

df = df.fillna('')
df['overview'] = (
    "Title: " + df["title"] +
    ". Genres: " + df["listed_in"] +
    ". Cast: " + df["cast"] +
    ". Description: " + df["description"]
)
```

```python
df['overview']
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')

```

```python
embeddings = model.encode(df['overview'].tolist())
```

```python
print(embeddings.shape)
similarities = model.similarity(embeddings, embeddings)
print(similarities)
```

```python
client = chromadb.PersistentClient(path="./chromaDB")

collection = client.get_or_create_collection("netflix_movies")
```

```python
for i in range(len(df)):
  collection.add(
      ids=[str(i)],
      embeddings=[embeddings[i].tolist()],
      documents=[df.iloc[i]['overview']],
      metadatas=[{"title": df.iloc[i]['title']}]
  )



```

```python
def search_movies(query, n_results=5):

    query_embedding = model.encode([query])

    results = collection.query(
        query_embeddings=query_embedding.tolist(),
        n_results=n_results
    )
```
